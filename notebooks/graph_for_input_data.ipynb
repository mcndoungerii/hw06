```python
import sys
from pathlib import Path
from data.input_data import DatasetCreator
import matplotlib.pyplot as plt
```

```python
# Get the absolute path of the current file
current_file_path = Path('graph_for_input_data.ipynb').resolve()

# Get the directory of the current file
project_dir = current_file_path.parent

# Add the project directory to sys.path
sys.path.insert(0, str(project_dir))


# Step 1: Create Datasets
dataset_creator = DatasetCreator()
blob_dataset = dataset_creator.create_blob_dataset()
points_dataset = dataset_creator.create_points_dataset()
```

```python
X_blob, y_blob = blob_dataset['X'], blob_dataset['y']
X_points = points_dataset['X']
```

```python
# Plot the data from make_blobs
plt.figure(figsize=(6, 4))
plt.scatter(X_blob[:, 0], X_blob[:, 1], c=y_blob, cmap='viridis', marker='o', edgecolor='k', s=50)
plt.title('Data generated by make_blobs')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.grid(True)
plt.show()
```

```python
# Plot the data for points per cluster
plt.figure(figsize=(6, 4))
plt.scatter(X_points[:, 0], X_points[:, 1], c='blue', marker='o', edgecolor='k', s=1)
plt.title('Manually generated clusters')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.grid(True)
plt.show()
```
